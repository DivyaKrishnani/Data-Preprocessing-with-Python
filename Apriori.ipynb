```python
import math
records = []  
"""
for i in range(0, df.shape[0]):
        for j in range(0, df.shape[1]):
            if(str(df.values[i,j]) != 'nan'):
                records.append([str(df.values[i,j])])
"""

for i in range(0, df.shape[0]):  
    records.append([str(df.values[i,j]) for j in range(0, df.shape[1])])
```

```python

association_rules = apriori(records, min_support=0.0045, min_confidence=0.20, min_lift=3, min_length=2)  
#association_rules = apriori(records)
association_rules = list(association_rules)  
```
