### Creating Sub Samples with Replacement

### Creating Sub Samples without Replacement

### Creating 10 Sub Samples of the above Sub Sample

```python
samplelist = []
for i in range(10):
    samplelist.append(sub_samples_wr.sample(n=10,replace=True))
    print(samplelist[i])
```

### Calculating measures of central tendency

Mean and Median for all numerical features and mode for all categorical features

```python
import statistics
numeric_features = ['age','bmi','children','expenses']
categorical_features = ['sex','smoker','region']
```

```python
mean10 = []
for i in range(10):
    mean10.append([samplelist[i][numeric_features[0]].mean(),samplelist[i][numeric_features[1]].mean(),samplelist[i][numeric_features[2]].mean(),samplelist[i][numeric_features[3]].mean()])
    
print(mean10)
print()
print(statistics.mean(data['age']))
print(statistics.mean(data['bmi']))
print(statistics.mean(data['children']))
print(statistics.mean(data['expenses']))
```

```python
median10 = []
for i in range(10):
    median10.append([samplelist[i][numeric_features[0]].median(),samplelist[i][numeric_features[1]].median(),samplelist[i][numeric_features[2]].median(),samplelist[i][numeric_features[3]].median()])
    
print(median10)
```

```python
mode10 = []
for i in range(10):
    mode10.append([samplelist[i][categorical_features[0]].mode(),samplelist[i][categorical_features[1]].mode(),samplelist[i][categorical_features[2]].mode()])
    
print(mode10)
```

### Calculating Dispersion

Variance, Standard Deviation, Range and IQR(Interquartile range)

```python
statistics.variance(samplelist[i][numeric_features[0]])
```

```python
var10 = []
for i in range(10):
    var10.append([statistics.variance(samplelist[i][numeric_features[0]]),statistics.variance(samplelist[i][numeric_features[1]]),statistics.variance(samplelist[i][numeric_features[2]]),statistics.variance(samplelist[i][numeric_features[3]])])
    
print(var10)
```

```python
stddev10 = []
for i in range(10):
    stddev10.append([statistics.stdev(samplelist[i][numeric_features[0]]),statistics.stdev(samplelist[i][numeric_features[1]]),statistics.stdev(samplelist[i][numeric_features[2]]),statistics.stdev(samplelist[i][numeric_features[3]])])
    
print(stddev10)

print()
print(statistics.stdev(data['age']))
print(statistics.stdev(data['bmi']))
print(statistics.stdev(data['children']))
print(statistics.stdev(data['expenses']))
```
